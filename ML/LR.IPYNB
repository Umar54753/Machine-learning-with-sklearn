```python
df_linear=df_linear[(df_linear['fare']>1)&(df_linear['fare']<30)]
```

```python
sns.boxplot(df_linear['age'])
```

```python
X=df_linear[['age',]]
Y=df_linear['fare']
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
model=LinearRegression()
model.fit(X,Y)
model.predict([[25]])
```

```python
X_train, X_test, Y_train,Y_test= train_test_split(X,Y,train_size=0.8)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)
prediction=model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
```

```python
MAE=mean_absolute_error(y_true=Y_test,y_pred=prediction)
MSE=mean_squared_error(y_true=Y_test,y_pred=prediction)
RMSE=root_mean_squared_error(y_true=Y_test,y_pred=prediction)
R2=r2_score(y_true=Y_test,y_pred=prediction)
```

```python
print(MAE)
print(MSE)
print(RMSE)
print(R2)
```
